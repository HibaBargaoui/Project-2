**Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Data**

```python
df = pd.read_csv('/content/MyDrive/MyDrive/Coding Dojo/Datasets/FMCG_data.csv')
df.head()
```

**Delete unnecessary columns**

```python
df = df.drop(columns = ['Ware_house_ID', 'WH_Manager_ID'])
df.head()
```

**Delete duplicate rows**

```python
# Check for duplicates
df.duplicated().sum()
```

There is no duplicate rows.

**Identify missing values**

```python
df.isna().sum()
```

**Adress missing values**

```python
# Adress the 'approved_wh_govt_certificate' column
df['approved_wh_govt_certificate'].value_counts()
```

```python
# Fill the 'approved_wh_govt_certificate' with the most frequent value
df['approved_wh_govt_certificate'].fillna(value='C',inplace=True)
```

```python
# Fill the 'workers_num' with the mean
df['workers_num'].fillna(value =df['workers_num'].mean(),inplace=True)
```

```python
# Fill the 'wh_est_year' with the mean
df['wh_est_year'].fillna(value =df['wh_est_year'].mean(),inplace=True)
```

```python
# Check for missing values
df.isna().sum()
```

Identify inconsistencies in data for categorical values

```python
cols = ['Location_type', 'WH_capacity_size', 'zone', 'WH_regional_zone', 'wh_owner_type', 'approved_wh_govt_certificate']
for col in cols:
  values = df[col].value_counts()
  print(f'{col}:\n{values}')
```

There are no inconsistencies in data for categorical values

**Correlations**

```python
# Identifying any moderate or strong correlations
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(10, 10))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu");
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
# Check the correlations
df.drop('product_wg_ton', axis=1).corrwith(df.product_wg_ton).plot(kind='bar', grid=True, figsize=(12, 8),
                                                   title="Correlation with target")
```

*Interpretation:*

'wh_est_year', 'storage_issue_reported_l3m' and 'wh_breakdown_l3m' are the most corrolated with the target.

**Multivariate visualizations**

```python
sns.lineplot(data=df, x='wh_est_year', y='product_wg_ton')
plt.xlabel('Warehouse_established_year')
plt.ylabel('Product Weight (tons)')
plt.title('Product Weight by year')
plt.show()
```

*Interpretation:*

Every year, the number of warehouses increases, but the total weight of shipped products decreases. For example, in the year 2020, we observed that less than 10,000 tons of products had been shipped.

```python
sns.lmplot(data=df, x='storage_issue_reported_l3m', y='product_wg_ton', hue='zone')
plt.xlabel('Storage issue in the last three months')
plt.ylabel('Product Weight (tons)')
plt.title('Product Weight by storage issues in the last three months')
plt.show()
```

*Interpretation:*

In various zones, as the quantity of products increases, we tend to encounter more storage-related issues due to rats, moisture and so on. Over the past three months, we've encountered 40 such storage issues.
